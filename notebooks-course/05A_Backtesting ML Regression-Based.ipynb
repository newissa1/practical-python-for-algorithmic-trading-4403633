## Load the data

```python
import pandas as pd

df = pd.read_excel('data/Microsoft_LinkedIn_Processed.xlsx', index_col=0, parse_dates=['Date'])
df
```

## Backtesting.py Library

### Create your Strategy Class

```python
from backtesting import Backtest, Strategy
```

```python
class Regression(Strategy):
    def init(self):
        self.model = model_dt
        self.already_bought = False

    def next(self):
        explanatory_today = pd.DataFrame([self.data.df.iloc[-1]])
        explanatory_today = explanatory_today[model_dt.feature_names_in_]  # align columns
        forecast_tomorrow = model_dt.predict(explanatory_today)[0]
        
        # conditions to buy or sell
        if forecast_tomorrow > 1 and self.already_bought == False:
            self.buy
            self.already_bought == True 
        elif forecast_tomorrow < -5 and self.already_bought == True:
            self.sell
            self.already_bought == False
        else:
            pass   
```

### Define initial conditions

```python
df_explanatory = df.drop(columns=['change_tomorrow','change_tomorrow_direction'])
```

```python
bt = Backtest(df_explanatory, Regression,
              cash=10000, commission=.002, exclusive_orders=True)
```

### Run backtesting

```python
results = bt.run()
```

### Interpret backtesting results

```python
results.to_frame(name='Values').loc[:'Return [%]']
```

### Visualize the strategy performance

```python
bt.plot(filename='reports_backtesting/classification.html')
```
